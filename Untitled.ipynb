```python
import numpy as np
```

```python
t = np.linspace(0, 10, 51)
```

```python
f = np.cos(t)
```

```python
plt.plot(t, f)
plt.xlabel(" t")
plt.ylabel(" f")
plt.suptitle(' f(t)')
plt.ylim((-2.5, 2.5))
plt.xlim((0.5, 9.5))
plt.show()
```

```python
x = np.linspace(-3, 3, 51)
```

```python
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(2, 2)
ax[0][0].plot(x, y1)
ax[0][0].set(title=" 1")
ax[0][0].set_xlim((-5, 5))
ax[0][1].plot(x, y2)
ax[0][1].set(title=" 2")
ax[1][0].plot(x, y3)
ax[1][0].set(title=" 3")
ax[1][1].plot(x, y4)
ax[1][1].set(title=" 4")
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```
